```python
import matplotlib.pyplot as plt
from ray_tracing_synthesis import SoSToTimeShiftTransformer, image_to_sos_map
from PIL import Image , ImageFilter 
import time
```

### Convert natural image to SoS map

```python
image_path = './example.jpg'  ## You may replace the exmaple image with your own image

# === Run ===
image = Image.open(image_path)
display(image)

sos_map = image_to_sos_map(image)

plt.imshow(sos_map,cmap = 'hot',vmax = 1600,vmin = 1400)
plt.title('SoS map')
plt.colorbar(label = '[m/s]')
```

### Derive time shift map from SoS map

```python
# Instantiate the class
transformer = SoSToTimeShiftTransformer()
```

```python
## Compute time shift map
time_shift_0 = transformer.transform(sos_map, '0psf')
time_shift_7p5 = transformer.transform(sos_map, '7p5psf')
time_shift_minus7p5 = transformer.transform(sos_map, 'minus7p5psf')
plt.imshow(time_shift_7p5)
plt.title('Time shift map')
plt.colorbar(label = '[s]')
```

```python
# Measure computation time for a time-shift map
start = time.time()
time_shift_minus7p5 = transformer.transform(sos_map, 'minus7p5psf')
end = time.time()
print(f"Time-shift map computed in {end - start:.6f} seconds")
```
